```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
mta_df = pd.read_csv("./Turnstile_Usage_Data__2017.csv")
```

```python
mta_df.sort_values(['C/A', 'Unit', 'SCP'])
```

```python
mta_df['datatime'] = pd.to_datetime(mta_df['Date'] + mta_df['Time'], 
                                  format='%m/%d/%Y%H:%M:%S')
```

```python
mta_df['Entries_shift_down'] = mta_df['Entries'].shift(1)
```

```python
mta_df['Entries_4_hours'] = mta_df['Entries'] - mta_df['Entries_shift_down']
```

```python
mta_df['Entries_4_hours'].fillna(0).describe()
```
